```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from pathlib import Path
```

```python
census = pd.read_csv("C:/Users/m8rqu/Documents/Data BootCamp/Opoid/Resources/va_2015.csv", encoding='utf-8')
census.head()
```

```python
# Replace removing (X)
census.columns = census.columns.str.strip()
census = census.replace('(X)', '')
census = census.dropna(axis =1, how='all')
count_of_empty = (census == '').all().sum()
print(count_of_empty)
```

```python
#Getting rid of rows that have margin, error percent
census = census[~census['Label (Grouping)'].str.contains('Margin|Error|Percent')]
census
```

```python
# Getting rows that only contain virgiina

virginia = census[census['Label (Grouping)'].str.contains('Virginia')]
virginia = virginia[~virginia['Label (Grouping)'].str.contains('West Virginia')]
virginia
```

```python
#Removing the Nas
virginia_drop = virginia.dropna(axis = 'columns')

# Replace 'N' with zero
virginia_drop_n = virginia_drop.replace('N', 0)

count_of_N = (virginia_drop_n == 'N').sum().sum()
print(count_of_N)
```

```python
# Define the file path where you want to save the CSV file
file_path = 'C:/Users/m8rqu/Documents/Data BootCamp/Opoid/Resources/virginia_2015.csv'

# Export the DataFrame to a CSV file with UTF-8 encoding
virginia_drop_n.to_csv(file_path, index=True, encoding='utf-8')
```

```python
# I can't believe that actually worked, kind of inefficent but whatever

va_data = pd.read_csv('C:/Users/m8rqu/Documents/Data BootCamp/Opoid/Resources/virginia_2015.csv')
va_data.dropna(axis =1 , inplace =True)
va_data
```

```python
del va_data['Unnamed: 0']

va_data = va_data.loc[:, va_data.any()]
va_data
```

```python
#Separating Datasets by Gender and Total

male_va = va_data['Label (Grouping)'].str.contains('!!Males!!Estimate')
female_va = va_data['Label (Grouping)'].str.contains("!!Females!!Estimate")
total_va = va_data['Label (Grouping)'].str.contains("!!Total!!Estimate")
```

```python
# Filter the original dataframe into separate dataframes for male and female estimates

male_df = va_data[male_va]
female_df = va_data[female_va]
total_df = va_data[total_va]
```

```python
# Split the 'data' column into separate columns for each dataframe
male_df[['state', 'gender', 'estimate']] = male_df['Label (Grouping)'].str.split('!!', expand=True)
female_df[['state', 'gender', 'estimate']] = female_df['Label (Grouping)'].str.split('!!', expand=True)
total_df[['state', 'gender', 'estimate']] = total_df['Label (Grouping)'].str.split('!!', expand=True)
```

```python
total_df.head()
```

```python
# Drop the original 'data' column from each dataframe
total_df.drop(['Label (Grouping)','gender','estimate'], axis =1, inplace = True)
```

```python
total_df.head()
```

```python
total_df['County'] = total_df['state'].str.split(', ').str[0]
total_df['Counties'] = total_df['County'].str.split(' ').str[0]
total_df['Counties'] = total_df['Counties'].str.upper()

```

```python
total_df.drop(['state','County'], axis =1, inplace =True)
```

```python
# Define the file path where you want to save the CSV file
file_path = 'C:/Users/m8rqu/Documents/Data BootCamp/Opoid/Resources/improved_2015.csv'

# Export the DataFrame to a CSV file with UTF-8 encoding
total_df.to_csv(file_path, index=False, encoding='utf-8')
```
